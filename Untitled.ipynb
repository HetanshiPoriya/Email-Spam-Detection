```python
# Text will be altered in the following ways :
# 1. Lower case
# 2. Tokenization
# 3. Remove special characters
# 4. Remove stop words and punctuation
# 5. Lemmatization
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
import string
string.punctuation
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lm = WordNetLemmatizer()
```

```python
# Our function for text tranformation
def transform_text(text) :

  # Convert to lower case
  text = text.lower()

  # Tokenization
  text = nltk.word_tokenize(text)

  # Removing special characters
  y = []
  for i in text :
    if i.isalnum() :
      y.append(i)

  # Removing stopwords and punctuations
  text = y[:]
  y.clear()

  for i in text :
    if i not in stopwords.words('english') and i not in string.punctuation :
      y.append(i)

  # lemmatizing
  text = y[:]
  y.clear()

  y = [lm.lemmatize(word) for word in text]

  # Return as a string
  return " ".join(y)
```

```python
from sklearn.model_selection import train_test_split

# X is the cleaned text data, y is the label
X = df['transformed_msg']
y = df['label'].values

# Split the *text* data first. Use a new random state to ensure a fresh split.
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

print(f"Training instances: {len(X_train_raw)}")
print(f"Testing instances: {len(X_test_raw)}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

# 1. FIT and TRANSFORM the training data
X_train_vectorized = cv.fit_transform(X_train_raw)

# 2. TRANSFORM the test data using the vocabulary learned from the training data (NO .fit())
X_test_vectorized = cv.transform(X_test_raw)

print(f"Vocabulary size (features): {X_train_vectorized.shape[1]}")
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif

# Initialize SelectKBest
k_selector = SelectKBest(mutual_info_classif, k=100)

# 1. FIT and TRANSFORM the training data
X_train_final = k_selector.fit_transform(X_train_vectorized, y_train)

# 2. TRANSFORM the test data using the features learned from the training data (NO .fit())
X_test_final = k_selector.transform(X_test_vectorized)

print(f"Final Training Data Shape: {X_train_final.shape}")
print(f"Final Testing Data Shape: {X_test_final.shape}")
```

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score

# Initialize models
mnb = MultinomialNB()
bnb = BernoulliNB()

# Train and Predict for Multinomial Naive Bayes
mnb.fit(X_train_final, y_train)
y_pred1_fixed = mnb.predict(X_test_final)

# Train and Predict for Bernoulli Naive Bayes
bnb.fit(X_train_final, y_train)
y_pred2_fixed = bnb.predict(X_test_final)

# Print Final Metrics
print("--- Multinomial Naive Bayes (Fixed) ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred1_fixed):.4f}")
print(f"Precision: {precision_score(y_test, y_pred1_fixed):.4f}")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred1_fixed))

print("\n--- Bernoulli Naive Bayes (Fixed) ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred2_fixed):.4f}")
print(f"Precision: {precision_score(y_test, y_pred2_fixed):.4f}")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred2_fixed))
```
